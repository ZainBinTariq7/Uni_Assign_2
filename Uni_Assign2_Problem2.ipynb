```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the Parquet file
# Replace 'your_local_file_path.parquet' with the downloaded file path
file_path = '/content/exercises_problem2.parquet'
data = pd.read_parquet(file_path)
```

```python
# Step 2: Explore the dataset
print(data.info())
print(data.describe())
print(data.head())
```

```python
# Step 3: Create visualizations

# Plot 1: Distribution of all numerical columns
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True)
plt.title('Distribution of Numerical Features', fontsize=16)
plt.xlabel('Values', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot 2: Correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

```python
# Plot 3: Pairplot for feature relationships
sns.pairplot(data)
plt.suptitle('Pairplot of Features', fontsize=16)
plt.show()
```

```python
# Plot 4: Boxplot for outlier detection
plt.figure(figsize=(12, 6))
sns.boxplot(data=data)
plt.title('Boxplot of Numerical Features', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

```python
# import modules
import numpy as nump
import matplotlib.pyplot as pyplot
```

```python
# Step-1: Generate 300 random numbers using normal Gaussian distribution
data = nump.random.normal(loc=5, scale=1.5, size=300)
```

```python
# Step-2: Calculate mean and median
meanValue = nump.mean(data)
medianValue = nump.median(data)
```

```python
# Step-3: Make histogram
pyplot.figure(figsize=(10, 6))
pyplot.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
```

```python
# Step-4: Annotate mean and median
pyplot.axvline(meanValue, color='red', linestyle='--', linewidth=2, label=f'Mean: {meanValue:.2f}')
pyplot.axvline(medianValue, color='green', linestyle='-', linewidth=2, label=f'Median: {medianValue:.2f}')
```

```python
pyplot.hist(data, bins=30, alpha=0.7, label='Random Data')
# Add style and presentation
pyplot.title('Histogram of Random Data (Normal Distribution)', fontsize=16)
pyplot.xlabel('Value', fontsize=14)
pyplot.ylabel('Frequency', fontsize=14)
pyplot.legend(fontsize=12)  # This will display the legend now
pyplot.xticks(fontsize=12)
pyplot.yticks(fontsize=12)
pyplot.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# Display the plot
pyplot.show()
```
